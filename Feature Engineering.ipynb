```python
import pandas as pd
data=pd.read_csv('adults.csv')
```

```python
data.loc[data['age'] <= 20, 'age'] = 0
data.loc[(data['age'] > 20) & (data['age'] <= 30), 'age'] = 1
data.loc[(data['age'] > 30) & (data['age'] <= 40), 'age'] = 2
data.loc[data['age'] > 40, 'age'] = 3
data.head()
```

```python
def category_ages(age):
 if age <= 20:
   return '<20 ans'
 elif (age > 20) & (age <= 30):
   return '20-30 ans'
 elif (age > 30) & (age <= 40):
   return '30-40 ans'
 else:
   return '+40 ans'
```

```python
data['age'].map(category_ages)
```
